```python
x=np.array([[1.1,0.1],[1.9,3.1],[1.8,0.8]])
y=np.array([1,1,-1],dtype=int)
```

```python
w=np.array([-2.0,1.0])
b=2.0

x0=np.linspace(1,2,2)
x1=-(b+w[0]*x0)/w[1]
plt.plot(x0,x1)
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.plot(x0,x1)
plt.show()
```

```python
def plotline(w,b,plt):
    x0=np.linspace(1,2,2)
    x1=-(b+w[0]*x0)/w[1]
    plt.plot(x0,x1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
w=np.array([-2.0,1.0])
b=2.0
plotline(w,b,plt)
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
w=np.array([-1.8,1.0])
b=2.0
plotline(w,b,plt)
plt.show()
```

```python
def predict(x,w,b):
    if b+w.dot(x)>0:
        return 1
    else:
        return -1
```

```python
w=np.array([-2.0,1.0])
b=2.0
print(predict(x[0],w,b), y[0])
print(predict(x[1],w,b), y[1])
print(predict(x[2],w,b), y[2])
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
w=np.array([-2.0,1.0])
b=2.0
plotline(w,b,plt)
plt.show()
print(predict(x[0],w,b), predict(x[1],w,b), predict(x[2],w,b))
```
