```python
def plotline(w,b,plt):
    x0=np.linspace(1,2,2)
    x1=-(b+w[0]*x0)/w[1]
    plt.plot(x0,x1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
w=np.array([-2.0,1.0])
b=2.0
plotline(w,b,plt)
plt.show()
```

```python
def predict(x,w,b):
    if b+w.dot(x)>0:
        return 1
    else:
        return -1
```

```python
def update(x,y,w,b,eta):
    if predict(x,w,b)==y:
        pass
    else:
        w,b=w+eta*x*y,b+eta*y
    return w,b
```

```python
class Perceptron():
    def __init__(self,w,b,n_iter,eta):
        self.w=w
        self.b=b
        self.n_iter=n_iter
        self.eta=eta
    
    def fit(self,X,y):
        for i in range(self.n_iter):
            for j in range(y.shape[0]):
                if self.predict(X[j])==y[j]:
                    pass
                else:
                    self.w,self.b=self.w+self.eta*X[j]*y[j],self.b+self.eta*y[j]
                    
    def predict(self,x):
        if self.b+self.w.dot(x)>0:
            return 1
        else:
            return -1
    
    def acc(self,x,y):
        y_pred=np.array([self.predict(x[i],self.w,self.b) for i in range(5)])
        return (y==y_pred).sum()/y.shape[0]
```
